```python
import pandas as pd

file_path = R"C:\MAIN\Practice programs\Sarcasm Detection\raw.csv"  # your file name
df = pd.read_csv(file_path)


duplicates = df[df.duplicated(subset=["Tweet"], keep=False)]

# KEEPING ONLY DUPLICATES THAT HAVE DIFFERENT LABELS ===
varying_label_duplicates = duplicates.groupby("Tweet").filter(lambda x: x["Label"].nunique() > 1)

# REMOVING DUPLICATES (KEEPING LAST OCCURRENCE) ===
unique_df = df.drop_duplicates(subset=["Tweet"], keep='last')

# === 4️⃣ SAVE CLEAN & DUPLICATE DATASETS ===
unique_df.to_csv("unique_tweets.csv", index=False, encoding="utf-8")
varying_label_duplicates.to_csv("duplicates_varying_labels.csv", index=False, encoding="utf-8")

print("✅ Saved cleaned unique dataset as unique_tweets.csv")
print("✅ Saved duplicates with varying labels as duplicates_varying_labels.csv")

print(f"Total original: {len(df)}")
print(f"Unique after cleaning: {len(unique_df)}")
print(f"Duplicates with varying labels: {len(varying_label_duplicates)}")
print("\nSample of duplicates with varying labels:")
print(varying_label_duplicates.head(10))
```
